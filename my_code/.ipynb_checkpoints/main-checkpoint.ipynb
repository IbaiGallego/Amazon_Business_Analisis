```python
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np
```

```python
you = pd.read_csv('../data/youtube.csv', encoding='latin-1')
```

```python
you2 = you.dropna()
```

```python
you2.columns = you2.columns.str.replace('_','')
```

```python
you2.columns = you2.columns.str.replace(' ','')
```

```python
you2.columns = you2.columns.str.strip()
```

```python
you2.columns = you2.columns.str.lower()
```

```python
def cambiar_comas_por_puntos(df):
    
    def sustituir(value):
        
        return str(value.replace('.', ','))  
    
    for columna in df.columns:
        
        df[columna] = df[columna].astype(str)
        
        df[columna] = df[columna].apply(sustituir)
        
    return df
```

```python
you3 = cambiar_comas_por_puntos(you2)
```

```python
months = {'Jan':1, 'Feb':2, 'Mar':3, 'Apr':4, 'May':5, 'Jun':6, 'Jul':7, 'Aug':8, 'Sep':9, 'Oct':10, 'Nov':11, 'Dec':12}
you3['date'] = [f"{str(row['createdyear'].split(',')[0])}/{str(months[row['createdmonth']])}/{str(row['createddate'].split(',')[0])}" for _, row in you3.iterrows()]
```

```python
bad_cols=['createddate', 'rank', 'createdmonth', 'createdyear', 'abbreviation', 'channeltyperank','countryrank','videoviewsrank',]
you3.rename(columns={'grosstertiaryeducationenrollment(%)': 'education_country','subscribersforlast30days': 'subs30d','videoviewsforthelast30days': 'views30d'} , inplace=True )

you3.drop(columns=bad_cols, axis=1, inplace=True)
```

```python
you3.to_csv('../data/youtube_clean.csv')
```

```python
def type_change(df):
    
    def sustituir2(value):
        
        return str(value.replace(',', '.'))  
    
    for columna in df.columns:
        
        df[columna] = df[columna].astype(str)
        
        df[columna] = df[columna].apply(sustituir2)
        
        try:
            df[columna] = df[columna].astype(int)
        except:
            df[columna] = df[columna]
        
    return df
```

```python
you4 = type_change(you2)
```

```python
you4.videoviews = you4.videoviews.astype(float)
```

```python
you4.subscribers = you4.subscribers.astype(float)
you4.uploads = you4.uploads.astype(float)
```

```python
you4.lowestyearlyearnings = you4.lowestyearlyearnings.astype(float)
you4.highestyearlyearnings = you4.highestyearlyearnings.astype(float)
you4.uploads = you4.uploads.astype(float)
```

```python
country = you4.pivot_table(values=['subscribers', 'videoviews', 'uploads', 'views30d', 'lowestmonthlyearnings', 'highestmonthlyearnings', 'lowestyearlyearnings', 'highestyearlyearnings', 'subs30d', 'population', 'unemploymentrate'], index=['country'], aggfunc={'subscribers': 'sum', 'videoviews': 'sum', 'uploads': 'sum', 'views30d': 'sum', 'lowestmonthlyearnings': 'sum', 'highestmonthlyearnings': 'sum', 'lowestyearlyearnings': 'sum', 'highestyearlyearnings': 'sum', 'subs30d': 'sum', 'population': 'max', 'unemploymentrate': 'max'})
```

```python
country = you4.pivot_table(
    values=['subscribers', 'videoviews', 'uploads', 'views30d', 'lowestmonthlyearnings',
            'highestmonthlyearnings', 'lowestyearlyearnings', 'highestyearlyearnings',
            'subs30d'],
    index=['country'],
    aggfunc='sum'
)
```

```python

you_countries = country.reset_index()

you_countries = cambiar_comas_por_puntos(you_countries)
```

```python
you_countries.to_csv('../data/you_countries.csv', index=False)
```
